```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.drop(columns=['charges'])
y = df['charges']
```

```python
import numpy as np
```

```python
y_log = np.log1p(y) ## y_log = np.log(y + 1)
```

```python
y_log
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y_log, random_state = 42, test_size = 0.2)
```

```python
X_train
```

## Handling Outliers

```python
def remove_outliers(X_train, y_train, column):
    Q1 = X_train[column].quantile(0.25)
    Q3 = X_train[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    mask = (X_train[column] >= lower_bound) & (X_train[column] <= upper_bound)
    
    return X_train[mask], y_train[mask]
```

```python
X_train, y_train = remove_outliers(X_train, y_train, 'bmi')
```

## Encoder

```python
X_train_encoded = pd.get_dummies(X_train,drop_first=True,dtype=int)
X_test_encoded = pd.get_dummies(X_test,drop_first=True,dtype=int)
```

```python
X_test_encoded = X_test_encoded.reindex(columns=X_train_encoded.columns, fill_value=0)
```

```python
X_train_encoded
```

```python
X_test_encoded
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
num_features = ['age', 'bmi']
```

```python
X_train_encoded[num_features] = scaler.fit_transform(X_train_encoded[num_features])
```

```python
X_test_encoded[num_features] = scaler.transform(X_test_encoded[num_features])
```

```python
X_train_scaled = X_train_encoded.copy()
X_test_scaled = X_test_encoded.copy()
```

```python
X_train_scaled
```

```python
X_test_scaled
```

## Fitting Multiple Linear Regression Model

```python
lr_model = LinearRegression()
lr_model.fit(X_train_scaled, y_train)
```

```python
y_pred_lr = lr_model.predict(X_test_scaled)
```

```python
y_pred_lr = np.expm1(y_pred_lr)
y_pred_lr
```

```python
y_test
```

```python
y_test_original = np.expm1(y_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python

print("ðŸ”¹ Multiple Linear Regression:")
print(f"  - MAE: {mean_absolute_error(y_test_original, y_pred_lr):.2f}")
print(f"  - MSE: {mean_squared_error(y_test_original, y_pred_lr):.2f}")
print(f"  - RÂ² Score: {r2_score(y_test_original, y_pred_lr):.2f}")
```
